```python
class Bird():
    def __init__(self,Name_EN,Name_LA,Freq):
        self.Name_En=Name_EN
        self.Name_La=Name_LA
        self.Name_Fr=None
        self.Freq_week=Freq
        self.__comput_freq()
        self.taxon=None
        self.family=None
        
    def __str__(self):
        string="Name: " + self.Name_En + "\t" + self.Name_La
        if self.Name_Fr:
            string += "\t" + self.Name_Fr
        string += "\t" + str(self.Freq_year)
        string += "\t" + str(self.taxon)
        string += "\t" + self.family
        return string
    
    def __comp__(self,other):
        if self.taxon < other.taxon : return -1
        elif self.taxon > other.taxon : return 1
        else: return 0
    def __lt__(self, other):
        return self.taxon < other.taxon
    def __gt__(self, other):
        return self.taxon > other.taxon
    def __eq__(self, other):
            return self.taxon == other.taxon
    def __le__(self, other):
         return self.taxon <= other.taxon
    def __ge__(self, other):
        return self.taxon >= other.taxon
    def __ne__(self, other):
        return self.taxon != other.taxon
        
    
    def __comput_freq(self):
        # Month
        t_m = []
        self.Freq_month=[]
        for i in range(0,len(self.Freq_week)):
            t_m.append(self.Freq_week[i])
            if i%4==3:
                self.Freq_month.append(reduce(lambda x, y: x + y, t_m) / len(t_m))
                t_m=[]
        assert len(self.Freq_month)==12,'Month Frequency is not equal to 12'
        #Season
        self.Freq_season=[]
        self.Freq_season.append(reduce(lambda x, y: x + y, self.Freq_month[2:5]) / 3)
        self.Freq_season.append(reduce(lambda x, y: x + y, self.Freq_month[5:8]) / 3)
        self.Freq_season.append(reduce(lambda x, y: x + y, self.Freq_month[8:11]) / 3)
        self.Freq_season.append(reduce(lambda x, y: x + y, self.Freq_month[0:2]+ [self.Freq_month[11]]) / 3)
        #Year
        self.Freq_year =reduce(lambda x, y: x + y, self.Freq_week) / len(self.Freq_week)
      
        
    def addLang(self,French_name):
        if French_name:
            French_name=French_name[0][0]
        self.Name_Fr = French_name
        
    def addinfo(self, taxon, family):
        if taxon:
            taxon=taxon[0][0]
        if family:
            family=family[0][0]
        self.taxon=float(taxon.replace(',','.'))
        self.family=family
```

```python
def AddLanguage(Birdlist,filename):
    import csv
    import sqlite3

    db = sqlite3.connect(':memory:')

    def init_db(cur):
        cur.execute('''CREATE TABLE Multilangue (
            Scientific_name TEXT,
            English TEXT,
            Chinese TEXT,
            Chinese_Traditional TEXT,
            Czech TEXT,
            Danish TEXT,
            Dutch TEXT,
            Estonian TEXT,
            Finnish TEXT,
            French TEXT,
            German TEXT,
            Hungarian TEXT,
            Italian TEXT,
            Japanese TEXT,
            Lithuanian TEXT,
            Norwegian TEXT,
            Polish TEXT,
            Portuguese TEXT,
            Russian TEXT,
            Slovak TEXT,
            Spanish TEXT,
            Swedish TEXT)''')

    def populate_db(cur, csv_fp):
        rdr = csv.reader(csv_fp)
        cur.executemany('''
            INSERT INTO Multilangue (Scientific_name,English,Chinese,Chinese_Traditional,Czech,Danish,Dutch,Estonian,Finnish,French,German,Hungarian,Italian,Japanese,Lithuanian,Norwegian,Polish,Portuguese,Russian,Slovak,Spanish,Swedish)
            VALUES (?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?)''', rdr)

    cur = db.cursor()
    init_db(cur)
    populate_db(cur, open(filename))
    db.commit()

    for bird in Birdlist:
        t=tuple([bird.Name_La],)
        cur.execute('SELECT French FROM Multilangue WHERE Scientific_name=?', t)
        bird.addLang(cur.fetchall())

    db.close()
```

```python
def AddInfo(Birdlist,filename):
    import csv
    import sqlite3

    db = sqlite3.connect(':memory:')

    def init_db(cur):
        cur.execute('''CREATE TABLE Taxonomy (
           TAXON_ORDER TEXT,
           CATEGORY TEXT,
           SPECIES_CODE TEXT,
           SCI_NAME TEXT,
           PRIMARY_COM_NAME TEXT,
           ORDER1 TEXT,
           FAMILY TEXT,
           SPECIES_GROUP TEXT,
           REPORT_AS TEXT)''')

    def populate_db(cur, csv_fp):
        rdr = csv.reader(csv_fp)
        cur.executemany('''
            INSERT INTO Taxonomy (TAXON_ORDER,CATEGORY,SPECIES_CODE,SCI_NAME,PRIMARY_COM_NAME,ORDER1,FAMILY,SPECIES_GROUP,REPORT_AS)
            VALUES (?,?,?,?,?,?,?,?,?)''', rdr)

    cur = db.cursor()
    init_db(cur)
    populate_db(cur, open(filename))
    db.commit()


    for bird in Birdlist:
        t=tuple([bird.Name_La],)
        cur.execute('SELECT TAXON_ORDER FROM Taxonomy WHERE SCI_NAME=?', t)
        taxon=cur.fetchall()
        cur.execute('SELECT FAMILY FROM Taxonomy WHERE SCI_NAME=?', t)
        family=cur.fetchall()
        bird.addinfo(taxon, family)

    db.close()
```

```python
class TableInput():
    def __init__(self,typ,option1=None,option2=None):
        self.typ=typ
        self.title=None
        self.wid=None
        self.option1=option1
        self.option2=option2
        if self.typ=='Language':
            self.wid=1
            if self.option1 == 'EN':
                self.title='English'
            elif self.option1 == 'LA':
                self.title='Latin'
            elif self.option1 == 'FR':
                self.title='Francais'
                
        elif self.typ=='Freq':
            self.wid=.4
            if self.option1=='Year':
                self.title='Year'
            elif self.option1 =='Season':
                season=['Spring','Summer','Fall','Winter']
                self.title=season[self.option2]
            elif self.option1 == 'Month':
                month=['Jan.', 'Feb.', 'Mar.', 'Apr.', 'May', 'June', 'Jul', 'Aug', 'Sept.', 'Oct.', 'Nov.', 'Dec.']
                self.title=month[self.option2]
        elif self.typ=='Note':
            if self.option1:
                self.wid=self.option1
            else:
                self.wid=1
            self.title='Note'
        elif self.typ=='Tiret':
            if not self.option1:
                self.option1=3
            if self.option2:
                self.wid=self.option1*self.opntion2
            else:
                self.wid=self.option1*0.1
            alphabet='A B C D E F G H'
            self.title=alphabet[:2*self.option1]
        
    def __str__(self):
        return str(self.title)
            
    def get_content(self,bird):
        if self.typ=='Language':
            self.wid=1
            if self.option1 == 'EN':
                return str(bird.Name_En)
            elif self.option1 == 'LA':
                return '\\textit{'+str(bird.Name_La)+'}'
            elif self.option1 == 'FR':
                if bird.Name_Fr:
                    return str(bird.Name_Fr)
                else:
                    return ''
        elif self.typ=='Freq':
            self.wid=.4
            if self.option1=='Year':
                return  '\\databar{'+'{:.1f}'.format(bird.Freq_year*100) +'}'
            elif self.option1 =='Season':
                return '\\databar{'+'{:.1f}'.format(bird.Freq_season[self.option2]*100) +'}'
            elif self.option1 == 'Month':
                return '\\databar{'+'{:.1f}'.format(bird.Freq_month[self.option2]*100) +'}'
        elif self.typ=='Note':
            return '\\dotfill'
        elif self.typ=='Tiret':
            if not self.option1:
                self.option1=3
            return self.option1*'\\underline{\\hspace{0.4cm}}'+' '
        
```

```python
def WriteToLatex(filename,Col,condition_tableau,condition_rare):
    family_current=''
    
    sum_wid=0.15*len(Col)
    for col in Col:
        sum_wid += col.wid

    # Start Writing
    f = open(filename, 'w')
    
    # Import preformatted text
    f2 = open('Latex/col_color.txt', 'r')
    for line in f2:
        if '\newcommand{\maxnum}{' in line:
            line += '70.00}'
        f.write(line)
    f2.close()
    
    for col in Col: # write column width
        f.write('m{' + str(col.wid/sum_wid) +'\\textwidth}')
    f.write('}\n')
    f.write('\\hline\n')

    for col in Col[:-1]: # write title
        f.write('\\textsc{ \\Large{'+col.title+'}} \t & ')
    f.write('\\textsc{ \\Large{ '+Col[-1].title+'}} \\\\[3ex] \n')
    f.write('\\hline\n')

    for bird in Birdlist: #write Content
        if eval(condition_tableau): # condition to display or not a bird
            if bird.family != family_current: # family name
                f.write('\\multicolumn{'+str(len(Col))+'}{c}{\\textbf{'+bird.family+'}} \\\\[2ex] \n')
                f.write('\\hline\n')
                family_current=bird.family

            for col in Col[:-1]:
                f.write(col.get_content(bird) + ' \t & ') # Content of the cell
            f.write(Col[-1].get_content(bird)+' \\\\ \n') # end of line
    
    f.write('\\hline \n')    #End Content
    f.write('\\end{longtable} \n')
    
    
    # Table Rare
    n_rare_col=4
    Col_r = TableInput('Language','EN')
    Birdlist_r=[]
    for bird in Birdlist:
        if eval(condition_rare):
            Birdlist_r.append(Col_r.get_content(bird))
           
    u=len(Birdlist_r)
    
    if u != 0: # check if there are rare bird
        while u-n_rare_col*round(u/n_rare_col)==0:
            Birdlist_r.append('')
            u=len(Birdlist_r)
        c=round(u/n_rare_col)
        f.write('\\begin{longtable}{')
        for x in range(0,n_rare_col):
            f.write('l p{' + str(1/(n_rare_col+2)) +'\\textwidth}')
        f.write('} \\\\ \n')
        f.write('\\hline\n')
        f.write('\\multicolumn{'+str(2*n_rare_col)+'}{c}{\\textbf{Rare}} \\\\ \n')
        f.write('\\hline\n')
        for cc in range(0,c):
            for c_r in range(0,n_rare_col-1):
                f.write('\_ \t &' + Birdlist_r[cc+c*c_r] +' \t &')
            f.write('\_ \t &' + Birdlist_r[cc+c*(n_rare_col-1)]+' \\\\ \n')
        f.write('\\hline\n')
        f.write('\\end{longtable} \n')
    f.write('\\end{center} \n\n')
    f.write('\\end{document}')

    f.close()
```
